```python
# 1.Create a new dataframe by filtering all the columns [i.e., features] except N_total, N_bulk, N_surface and R_avg columns.

import pandas as pd

data = pd.read_csv('C:\\Users\\Tech\\Desktop\\SLIIT\\4 - Fourth Year\\2023 July First Semester\\DL\\Lab\\1\\Au_nanoparticle_dataset.csv')

selected_columns = ['N_total', 'N_bulk', 'N_surface', 'R_avg']
filtered_data = data[selected_columns]

filtered_data.head()
```

```python
# 2.Display the first 20 samples of this dataframe.

print(filtered_data.head(20))
```

```python
# 3.Calculate the mean, standard deviation and quartile values for each of the above 4 features.

mean_values = filtered_data.mean()
std_values = filtered_data.std()
quartile_values = filtered_data.quantile([0.25, 0.5, 0.75])

for column in selected_columns:
    print(f"Column: {column}")
    print("Mean:", mean_values[column])
    print("Standard Deviation:", std_values[column])
    print("Quartiles:")
    print(quartile_values[column])
    print()
```

```python
# 4.Plot the histogram of each of these features in a 1x4 layout.

import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

for i, column in enumerate(selected_columns, 1):
    plt.subplot(1, 4, i)
    plt.hist(filtered_data[column], bins=20, edgecolor='black', alpha=0.7)
    plt.title(column)
    plt.xlabel('Value')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# 5.Visualize the scatter plots and histograms of this dataframe using the pairplot functionality of seaborn library.

import seaborn as sns


sns.set(style='ticks')
sns.pairplot(filtered_data)
plt.show()
```

```python
# 6.Add the below code and change it such that,
# a.Plots on the diagonal contains the histogram of each feature along with the kernal density estimation plot.
# b.Plots on the lower half [g.map_lower] contains the bivariate kernal density estimation plot.

g = sns.PairGrid(filtered_data)

g.map_upper(sns.histplot)

g.map_diag(sns.histplot, kde=True)

g.map_lower(sns.kdeplot)

plt.tight_layout()
plt.show()

```
