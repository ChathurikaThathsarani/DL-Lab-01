```python
#2.	Create a random 100000x1 array from the same exponential distribution and visualize the distribution histogram along with uniform and normal distributions. Use the code given below and do the necessary changes. Change the number of bins and the plot view to make the visualization better.
import numpy as np
import matplotlib.pyplot as plt

plt.hist(np.random.exponential(scale=1.0, size=100000), density=True, bins=50, histtype="step", color="purple", label="Exponential")
plt.hist(np.random.rand(100000), density=True, bins=50, histtype="step", color="red", label="Uniform")
plt.hist(np.random.normal(size=100000), density=True, bins=50, histtype="step", color="pink", label="Normal")

plt.axis([-2.5, 2.5, 0, 1.1])
plt.legend(loc="upper right")
plt.title("Random Distributions")
plt.xlabel("Value")
plt.ylabel("Density")
plt.show()
```

```python
# 3.Using matplotlib, plot the Z = X**2 + Y**2 in a 3D plot. Limit the x and y values to [-5, 5]. 

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

fig = plt.figure()
subplot = fig.add_subplot(111, projection='3d')


subplot.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=0.1)

subplot.set_xlabel('X')
subplot.set_ylabel('Y')
subplot.set_zlabel('Z')
subplot.set_title('3D Plot of Z = X^2 + Y^2')

plt.show()
```

```python
# 4.Calculate the pearson standard coefficient and spearman rank correlation for the [HP, Attack, Defense, Sp. Ark, Sp. Def, Speed] features in the seaborn tutorial.ipynb. Visualize the results using heatmaps. Make sure the values are displaced in the heatmap as well.

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

path = 'C:\\Users\\Tech\\Desktop\\SLIIT\\4 - Fourth Year\\2023 July First Semester\\DL\\Lab\\1\\Pokemon.csv'
data = pd.read_csv(path, encoding='latin1') 

features = ['HP', 'Attack', 'Defense', 'Sp. Atk', 'Sp. Def', 'Speed']
filtered_data = data[features]

pearson_correlation = filtered_data.corr(method='pearson')
spearman_correlation = filtered_data.corr(method='spearman')

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(12, 6))

sns.heatmap(pearson_correlation, annot=True, fmt=".2f", cmap='coolwarm', ax=axes[0])
axes[0].set_title('Pearson Correlation')

sns.heatmap(spearman_correlation, annot=True, fmt=".2f", cmap='coolwarm', ax=axes[1])
axes[1].set_title('Spearman Correlation')

plt.tight_layout()

plt.show()
```
